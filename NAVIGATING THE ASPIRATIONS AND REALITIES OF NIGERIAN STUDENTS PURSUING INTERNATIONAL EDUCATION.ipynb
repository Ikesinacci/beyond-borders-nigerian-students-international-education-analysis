BEYOND BORDERS: NAVIGATING THE ASPIRATIONS AND REALITIES OF NIGERIAN STUDENTS PURSUING INTERNATIONAL EDUCATION

Problem Statement:
The purpose of this analysis is to explore the motivations, challenges, and realities faced by Nigerian students pursuing education abroad. Using data gathered from respondents during the The International Education (TINE) Africa Summit 2024 held at Sheraton Hotels, Ikeja, we aimed to uncover trends, key insights, and actionable findings that reflect largely, their expectations/aspirations and lived experiences.

IMPORTING LIBRARIES AND DEPENDENCIES (Integrating Survey Spreadsheets)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
sns.set(style="whitegrid")

# Defining and initializing Google Sheets
def load_google_sheet(sheet_url):
    sheet_id = sheet_url.split('/')[5]
    csv_url = f"https://docs.google.com/spreadsheets/d/{sheet_id}/export?format=csv"
    return pd.read_csv(csv_url)

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
```

LOADING SURVEY DATA

```python
student_data_url = "https://docs.google.com/spreadsheets/d/1DxnVb8Z__bt3--y5NjvNh7T_rl0XhXhW9runsD9yY40/edit?usp=sharing"
stakeholder_data_url = "https://docs.google.com/spreadsheets/d/1nragiZAFB8-yooegy4BFNnJywyXvwBTDcLTEVub942k/edit?usp=sharing"

student_df = load_google_sheet(student_data_url)
stakeholder_df = load_google_sheet(stakeholder_data_url)

print("Student Data:")
display(student_df.head())

print("Stakeholder Data:")
display(stakeholder_df.head())

student_df = student_df.dropna(how='all')  
stakeholder_df = stakeholder_df.dropna(how='all')
```

METHODOLOGY

1. Data Collection: 
   - Responses were gathered on the first day of the TINE Africa Summit and Awards Conference 2024. The participants primarily included Nigerian students aspiring to pursue international education, education agency representatives, professionals, and counselors within the international education sector. Data collection was conducted through structured surveys.
2. Data Cleaning:
   - Missing and redundant data were identified and addressed.
3. Analysis:
   - Key areas such as motivations, challenges, countries of interest, and funding options were explored.
4. Visualization:
   - Charts and graphs were utilized to visually present insights, enhancing clarity and comprehension. The preferred visualization tool for this purpose was Power BI.
5. Report:
   - The findings will highlight major aspirations, challenges, and the support required by Nigerian students in their international education journey.


Key Areas of Analysis:
1. Motivations for studying abroad
2. Challenges encountered
3. Preferred countries and institutions
4. Affordability of Institutions

1. Motivations for studying abroad

```python
print("Student Data Overview:")
print(student_df.info())

# Analyis of motivations for studying abroad
motivations_col = '1. What is motivating you to pursue international education?'  
motivations_count = student_df[motivations_col].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=motivations_count.index, y=motivations_count.values, palette="viridis")
plt.title("Motivations for Studying Abroad")
plt.xticks(rotation=45, ha="right")
plt.ylabel("Count")
plt.show()
```

2. Challenges encountered

```python
challenges_col = '6. What are your primary concerns about studying abroad?'
challenges_count = student_df[challenges_col].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=challenges_count.index, y=challenges_count.values, palette="rocket")
plt.title("Challenges Encountered by Nigerian Students Abroad")
plt.xticks(rotation=45, ha="right")
plt.ylabel("Count")
plt.show()
```

3. Preferred countries and institutions

```python
countries_col = '20. What is your preferred study destination?' 
countries_count = student_df[countries_col].value_counts()

plt.figure(figsize=(12, 6))
sns.barplot(x=countries_count.index, y=countries_count.values, palette="coolwarm")
plt.title("Top Countries Preferred for International Education")
plt.xticks(rotation=45, ha="right")
plt.ylabel("Count")
plt.show()
```

4. Affordability of Institutions

```python
funding_col = '9. How affordable do you consider the tuition fees for international education in your preferred study destination?' 
funding_count = student_df[funding_col].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=funding_count.index, y=funding_count.values, palette="Set2")
plt.title("Affordability of International Education")
plt.xticks(rotation=45, ha="right")
plt.ylabel("Count")
plt.show()
```

SUMMARY OF INSIGHTS

1. Motivations:
   - Most students are motivated by the prospect of career advancement.
2. Challenges:
   - The most predominant challenge cited by the students points to funding.
3. Preferred Countries:
   - The most popular destinations for international education are the UK, USA, and Canada.
4. Funding:
   - A significant proportion of students are convinced that international education is not affordable for the average Nigerian.

These insights highlight key focus areas for policymakers, educators, and stakeholders in international education.


LIMITATiONS

1. The Sample size may not be fully representative of all Nigerian students pursuing international education.
2. Responses are self-reported and may contain some degree of bias.
3. Limited data on post-study outcomes and experiences.

CONCLUSION

This analysis provides a comprehensive understanding of the aspirations and realities of Nigerian students pursuing overseas education.
